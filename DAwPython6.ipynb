# recap for Apply - Transform - Map

```python
np.random.seed(42)
df3 = pd.DataFrame(  np.random.randn(5,4), 
                   index = ['A', 'B', 'C', 'D', 'E'], 
                   columns = 'W X Y Z'.split()          )
df3
```

```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C','A', 'B', 'C'],
                   'var1': [10,23,33,22,11,99,40,20,15],
                   'var2': [100,253,333,262,111,969, 232,343,123]})
df6
```

```python
# df6[["var1","var2"]].transform(  )  # make minmaxscaler
df6[["var1"]].transform(lambda x : ( x - df6["var1"].min() )/
                                       (df6["var1"].max()-df6["var1"].min())
                       )
```

```python
# df6[["var1","var2"]].apply(  )  # make minmaxscaler
df6[["var1","var2"]].apply(lambda x : ( x - df6["var1"].min() )/
                                       (df6["var1"].max()-df6["var1"].min())  ) 
```

```python
# f6.groupby('groups')[["var1","var2"]].transform(np.mean)
df6.groupby('groups')[["var1","var2"]].transform(np.mean)
```

```python
# df6.groupby('groups')[["var1","var2"]].apply(np.mean)
```

```python
df6.groupby('groups')[["var1","var2"]].apply(np.mean)
```

# pivot table

```python
titanic = sns.load_dataset('titanic')
titanic                                  # 
```

##titanic

```python
titanic.describe()
```

```python
titanic.groupby(["who","sex","pclass",])[["survived"]].mean()
```

```python
titanic.pivot_table(values="survived", index="sex", columns="pclass")
```

```python
titanic.groupby(["sex","pclass",])[["survived"]].mean().unstack()
```

```python
titanic.groupby(["who","sex","pclass",])[["survived"]].mean()
```

```python
titanic.groupby(["who","sex","pclass",])[["survived"]].mean().unstack()
```

```python
titanic.groupby(["who","sex","pclass",])[["survived"]].mean().unstack().unstack()
```

```python
titanic.pivot_table(values="survived", index=["sex","who"], columns="pclass") #olusan rakamlar survived'in mean'leri
```

```python
titanic.pivot_table(values="survived", index=["sex","who"], columns="pclass",aggfunc="count") # aggfunc default mean iken biz count aldik
```

```python
titanic.pivot_table(values="survived", index=["sex","deck"], columns="pclass",aggfunc="count")
```

```python
titanic.groupby(["who","sex","pclass",])[["survived","fare"]].mean().unstack(1)
```
