```python
df11 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x', 'z'],
                    'lvalue': [2, 3, 5, 7, 0]})
df12 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
df11
```

```python
df12
```

```python
pd.merge(df11,df12, left_on="lkey", right_on="rkey", how = "inner") #ortak olanlari aldik inner olarak
```

```python
pd.merge(df11,df12, left_on="lkey", right_on="rkey", how = "left")
```

```python
pd.merge(df11,df12, left_on="lkey", right_on="rkey", how = "outer") #ikisinde de ortak olan olmayan ne var ne yok getiriyor
```

```python
df_left = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x'],
                    'lvalue': [2, 3, 5, 7]})
df_right = pd.DataFrame({'rkey': ['a', 'b', 'c', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
pd.merge(df_left,df_right, left_on="lkey", right_on="rkey", how = "inner") #ortak hicbir sey yok
```

```python
pd.merge(df_left,df_right, left_on="lkey", right_on="rkey", how = "left")
```

```python
pd.merge(df_left,df_right, left_on="lkey", right_on="rkey", how = "right")
```

```python
pd.merge(df_left,df_right, left_index=True, right_index=True, how = "outer")
```
